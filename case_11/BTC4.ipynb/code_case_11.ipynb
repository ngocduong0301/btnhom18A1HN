```python
import numpy as np
```

```python
def read_hr_csv(path):
    # Đọc file CSV và trả về các mảng: mã nhân viên, thâm niên, lương, điểm thăng tiến
    data = np.genfromtxt(path, delimiter=",", dtype=float, skip_header=1)
    if data.ndim == 1:
        data = data.reshape(1, -1)
    emp_id = data[:,0].astype(int)
    seniority = data[:,1].astype(float)
    salary = data[:,2].astype(float)
    promo = data[:,3].astype(float)
    return emp_id, seniority, salary, promo
```


# Đường dẫn tới file 


```python
file_a = "group_a.csv"
file_b = "group_b.csv"

emp_a, sen_a, sal_a, prom_a = read_hr_csv(file_a)
emp_b, sen_b, sal_b, prom_b = read_hr_csv(file_b)

group_a = np.array(["A"] * emp_a.shape[0])
group_b = np.array(["B"] * emp_b.shape[0])

employee_id = np.concatenate([emp_a, emp_b])
seniority = np.concatenate([sen_a, sen_b])
salary = np.concatenate([sal_a, sal_b])
promotion_score = np.concatenate([prom_a, prom_b])
group = np.concatenate([group_a, group_b])
```

# *Thống kê mô tả*

```python
def ThongKe(arr, name):
    print(f"\n=== {name} ===")
    print(f"Mean: {arr.mean():.2f} | Std: {arr.std():.2f} | Min: {arr.min():.2f} | Max: {arr.max():.2f}")

ThongKe(sen_a, "Thâm niên Group A")
ThongKe(sal_a,    "Lương Group A")
ThongKe(prom_a,     "Promotion Score Group A")

ThongKe(sen_b, "Thâm niên Group B")
ThongKe(sal_b,    "Lương Group B")
ThongKe(prom_b,     "Promotion Score Group B")
```


# Tính các tertile (33.33% và 66.67%) cho thâm niên

```python
ter1 = np.percentile(seniority, 33.3333)
ter2 = np.percentile(seniority, 66.6667)

low_mask = seniority < ter1
mid_mask = (seniority >= ter1) & (seniority < ter2)
high_mask = seniority >= ter2

low_mask_a = sen_a < ter1
low_mask_b = sen_b < ter1
mid_mask_a = (sen_a >= ter1) & (sen_a < ter2)
mid_mask_b = (sen_b >= ter1) & (sen_b < ter2)
high_mask_a = sen_a >= ter2
high_mask_b = sen_b >= ter2

print("\nSố lượng nhân viên theo thâm niên:")
print(f"Group A - Thấp: {low_mask_a.sum()}, Trung bình: {mid_mask_a.sum()}, Cao: {high_mask_a.sum()}")
print(f"Group B - Thấp: {low_mask_b.sum()}, Trung bình: {mid_mask_b.sum()}, Cao: {high_mask_b.sum()}")

```

# *PHÂN LOẠI TIỀM NĂNG & THU NHẬP*

```python
high_potential_a = prom_a >= 7.0
high_salary_a    = sal_a >= np.percentile(sal_a, 75)

high_potential_b = prom_b >= 7.0
high_salary_b    = sal_b >= np.percentile(sal_b, 75)

# Nhân viên tiềm năng: thâm niên cao + tiềm năng cao
star_a = high_mask_a & high_potential_a
star_b = high_mask_b & high_potential_b

print(f"\nNhân viên tiềm năng (thâm niên cao + promotion cao):")
print(f"Group A: {star_a.sum()} người | Group B: {star_b.sum()} người")

# Nhân viên cảnh báo: thâm niên cao nhưng promotion thấp (<4.0)
warning_a = high_mask_a & (prom_a < 4.0)
warning_b = high_mask_b & (prom_b < 4.0)
print(f"Nhân viên cảnh báo (thâm niên cao + promotion thấp):")
print(f"Group A: {warning_a.sum()} người | Group B: {warning_b.sum()} người")
```

# *TOP 10 LƯƠNG & TOP 10 PROMOTION*

```python
top10_salary_idx_a = np.argsort(sal_a)[-10:][::-1]
top10_promo_idx_a  = np.argsort(prom_a)[-10:][::-1]

top10_salary_idx_b = np.argsort(sal_b)[-10:][::-1]
top10_promo_idx_b  = np.argsort(prom_b)[-10:][::-1]

print("\nTop 10 lương cao nhất Group A:")
print("ID     Seniority  Salary   Promo")
for idx in top10_salary_idx_a[:5]:
    print(f"{emp_a[idx]:<6} {sen_a[idx]:6.1f}   {sal_a[idx]:7.0f}   {prom_a[idx]:.2f}")
print("\nTop 10 lương cao nhất Group B:")
print("ID     Seniority  Salary   Promo")
for idx in top10_salary_idx_b[:5]:
    print(f"{emp_b[idx]:<6} {sen_b[idx]:6.1f}   {sal_b[idx]:7.0f}   {prom_b[idx]:.2f}")
```


# Ngưỡng 75% cho điểm thăng tiến và lương

```python
promo_75 = np.percentile(promotion_score, 75)
sal_75 = np.percentile(salary, 75)
high_potential_mask = promotion_score >= promo_75
high_income_mask = salary >= sal_75
```

# Tạo mảng cấu trúc và lưu ra CSV với nhãn trường bằng tiếng Việt


```python
out_dtype = [
    ('ma_nv','i8'),
    ('nhom','U1'),
    ('thamnien','f8'),
    ('luong','f8'),
    ('diem_thang_tien','f8'),
    ('nhom_thamnien','U10'),
    ('tien_nang_cao','U5'),
    ('thu_nhap_cao','U5')
]
out_arr = np.zeros(employee_id.size, dtype=out_dtype)


seniority_group = np.full(employee_id.shape, "trung")
seniority_group[low_mask] = "thấp"
seniority_group[high_mask] = "cao"

out_arr['ma_nv'] = employee_id
out_arr['nhom'] = group
out_arr['thamnien'] = seniority
out_arr['luong'] = salary
out_arr['diem_thang_tien'] = promotion_score
out_arr['nhom_thamnien'] = seniority_group
out_arr['tien_nang_cao'] = np.where(high_potential_mask, "có", "không")
out_arr['thu_nhap_cao'] = np.where(high_income_mask, "có", "không")

print('ma_nv', 'nhom', 'thamnien', 'luong', 'diem_thang_tien', 'nhom_thamnien', 'tien_nang_cao', 'thu_nhap_cao', sep=', ')
for i in out_arr:
    print(i)
```

# in ra kết quả 

```python
with open("ket_qua_case_11.csv", "w", encoding="utf-8") as f:
    # Ghi header (tên trường sẽ bằng tiếng Việt như trên)
    f.write(",".join(out_arr.dtype.names) + "\n")
    for row in out_arr:
        line = (
            f"{int(row['ma_nv'])},{row['nhom']},{row['thamnien']:.6f},{row['luong']:.6f},"
            f"{row['diem_thang_tien']:.6f},{row['nhom_thamnien']},"
            f"{row['tien_nang_cao']},{row['thu_nhap_cao']}\n"
        )
        f.write(line)

print("Đã lưu ket_qua_case_11.csv")
```
